# Libraries

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Embedding
from tensorflow.keras.layers import Conv1D, GlobalMaxPooling1D
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
import numpy as np
import os
```

```python
archivo = open("C:/Users/jds25/Desktop/Hack a Boss/Proyecto 3/Proyecto-3/Data/Texts/Alice_in_wonderland.txt", "r", encoding= 'utf-8')
texts = archivo.read()  # Lee todo el contenido del archivo
archivo.close()  # Cierra el archivo despu√©s de usarlo
```

```python
# Tokenizaci√≥n
tokenizer = Tokenizer(num_words=10000) 
tokenizer.fit_on_texts(texts)
sequences = tokenizer.texts_to_sequences(texts)
print("Sequences:\n", sequences)

# Padding
max_length = 5 # Cantidad de palabras diferentes
padded_sequences = pad_sequences(sequences, maxlen=max_length)

print("Padded: \n", padded_sequences)
```

```python
embedding_layer = Embedding(input_dim=10000, output_dim=128, input_length=max_length)
# Capa Convolucional y Pooling
conv_layer = Conv1D(filters=128, kernel_size=5, activation='relu')
pooling_layer = GlobalMaxPooling1D() # GlobalMaxPooling1D Es la comunmente utilizada para tomar el valor maximo de la secuencia en texto
```

```python
model = Sequential([
    embedding_layer,
    conv_layer,
    pooling_layer,
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

# Compilar el modelo
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Resumen del modelo
model.summary()
```

```python
# Datos de ejemplo
x_train = padded_sequences
y_train = np.array([1, 0])   # 1 para positivo, 0 para negativo

# Entrenamiento del modelo
model.fit(x_train, y_train, epochs=5, batch_size=2)
```

```python
# Prueba del modelo
test_texts = ["I really love this movie.", "I really hate this movie."]
test_sequences = tokenizer.texts_to_sequences(test_texts)
test_padded_sequences = pad_sequences(test_sequences, maxlen=max_length)

# Obtener predicciones
predictions = model.predict(test_padded_sequences)
print("Predictions: \n", predictions)

# Interpreta las predicciones
for i, test_text in enumerate(test_texts):
    print(f"Text: {test_text}")
    print(f"Predicted sentiment: {'Positive' if predictions[i] > 0.5 else 'Negative'}")
```
